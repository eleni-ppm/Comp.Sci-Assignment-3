```python
# your code goes here
from functools import reduce

students = [
    {'name': 'Student 1', 'scores': [92, 60, 76, 84, 68, 66]},
 {'name': 'Student 2', 'scores': [79, 73, 58, 93, 67]},
 {'name': 'Student 3', 'scores': [99, 97, 53, 99, 83]}
]

def calculate_average(student):
    total_score = reduce(lambda x, y: x + y, student["scores"])
    average_score = total_score / len(student["scores"])
    student["average_score"] = average_score
    return student

students_with_avg = list(map(calculate_average, students))


students_above_90 = list(filter(lambda student: student["average_score"] > 90, students_with_avg))


print(students_above_90)

```

```python
# your code goes here
# hints: 1) Group products by category (you don't need to use map reduce for this part), then 2) use map reduce paradigm to
# calculate the average price for each category and filter categories with an average price > 50

products = [
    {"name": "Product 1", "price": 62, "category": "Books"},
    {"name": "Product 2", "price": 52, "category": "Electronics"},
    {"name": "Product 3", "price": 41, "category": "Home"},
    {"name": "Product 4", "price": 117, "category": "Home"},
    {"name": "Product 5", "price": 60, "category": "Electronics"}
]

from functools import reduce
from collections import defaultdict


def group_by_category(products):
    category_groups = defaultdict(list)
    for product in products:
        category_groups[product["category"]].append(product)
    return category_groups


def calculate_average_price(products_in_category):
    total_price = reduce(lambda x, y: x + y["price"], products_in_category, 0)
    average_price = total_price / len(products_in_category)
    return average_price


def filter_categories_by_avg_price(categories_with_avg):
    return [
        {"category": category, "average_price": avg_price}
        for category, avg_price in categories_with_avg.items() if avg_price >= 50
    ]


category_groups = group_by_category(products)


category_avg_prices = {
    category: calculate_average_price(products_in_category)
    for category, products_in_category in category_groups.items()
}


filtered_categories = filter_categories_by_avg_price(category_avg_prices)

# Output 
print(filtered_categories)


```

```python
# your code goes here
# hints: 1) Group employees' salaries by department (you don't need to use map reduce for this part), then 2) use map reduce paradigm to
# calculate the average salary for each department and filter departments with an average salary > threshold

from functools import reduce
from collections import defaultdict

employees = [
    {'name': 'Employee 1', 'salary': 92017, 'department': 'Finance'},
 {'name': 'Employee 2', 'salary': 64694, 'department': 'IT'},
 {'name': 'Employee 3', 'salary': 44054, 'department': 'Engineering'}
]

def group_by_department(employees):
    department_groups = defaultdict(list)
    for employee in employees:
        department_groups[employee["department"]].append(employee)
    return department_groups


def calculate_average_salary(employees_in_department):
    total_salary = reduce(lambda x, y: x + y["salary"], employees_in_department, 0)
    average_salary = total_salary / len(employees_in_department)
    return average_salary


def filter_departments_by_avg_salary(departments_with_avg):
    return [
        {"department": department, "average_salary": avg_salary}
        for department, avg_salary in departments_with_avg.items() if avg_salary > 65000
    ]


department_groups = group_by_department(employees)


department_avg_salaries = {
    department: calculate_average_salary(employees_in_department)
    for department, employees_in_department in department_groups.items()
}


filtered_departments = filter_departments_by_avg_salary(department_avg_salaries)


print(filtered_departments)

```

Write the following five functions to analyze global alignments between two sequences using Biopython's `pairwise2` module:

1. **countMatches(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment (pairwise2.globalxx) of the same length. It returns the number of positions where the elements of both sequences match.

2. **countMismatches(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment of the same length. It returns the number of positions where the elements of the two sequences are different (i.e., they are not gaps, and the characters do not match).

3. **countGapOpens(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment of the same length. It returns the number of gap openings in the alignment (a gap is opened when a '-' appears in the sequence).

4. **countGapExtensions(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment of the same length. It returns the number of gap extensions (where '-' continues in the alignment after an initial gap is opened).

5. **getScore(s1, s2, matchScore, mismatchPenalty, gapOpenPenalty, gapExtensionPenalty)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment and returns the alignment score based on the provided scoring scheme: `matchScore` for matches, `mismatchPenalty` for mismatches, `gapOpenPenalty` for opening a gap, and `gapExtensionPenalty` for extending a gap.
